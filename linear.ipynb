```python
# imports

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
```

```python
# Importar bases de dados

df = pd.read_csv('dados_parte08.csv')
```

```python
y = df['MEDV']

X = df.drop(['MEDV'],axis=1)

X_treino, X_teste, y_treino, y_teste = train_test_split(X,y, train_size=0.7)
```

```python
# Treinamento do modelo

model = LinearRegression()

model.fit(X_treino, y_treino)
```

```python
# Exibir os coeficientes do modelo

atributos = X.columns.tolist()

coeficientes = model.coef_.tolist()

ruido = model.intercept_


termos = []

for ai, bi in zip(coeficientes,atributos):

    termos.append(f"({ai:.2f}) * {bi}")


expressao = " + ".join(termos)+f" + ({ruido:.2f})"

print(expressao)
```

```python
# Predição da variável de resposta nos dados de teste

y_pred = model.predict(X_teste)

df_teste = X_teste

df_teste['y']=y_teste

df_teste['y_predito']=y_pred


print(df_teste)
```
