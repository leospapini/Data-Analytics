```python
# imports

from sklearn.ensemble import RandomForestClassifier

import pandas as pd
```

```python
# Criar DataFrames com dados de treino e teste

df_treino = pd.DataFrame({

	'Dia': [1, 2, 3, 4, 5, 6, 7, 8],

	'Tempo': ['Ensolarado', 'Nublado', 'Chuvoso', 'Ensolarado', 'Nublado', 'Chuvoso', 'Nublado', 'Chuvoso'],

	'Temperatura': ['Quente', 'Quente', 'Moderado', 'Moderado', 'Frio', 'Frio', 'Moderado', 'Quente'],

	'JogaFutebol': ['Sim', 'Sim', 'Não', 'Sim', 'Não', 'Não', 'Sim', 'Não']

})


df_teste = pd.DataFrame({

	'Dia': [9, 10],

	'Tempo': ['Chuvoso', 'Ensolarado'],

	'Temperatura': ['Quente', 'Frio']

})
```

```python
# Converter variáveis categóricas em numéricas

temperatura_map = {'Quente': 0, 'Moderado': 1, 'Frio': 2}

tempo_map = {'Chuvoso': 0, 'Nublado': 1, 'Ensolarado': 2}


df_treino['Temperatura'] = df_treino['Temperatura'].map(temperatura_map)

df_teste['Temperatura'] = df_teste['Temperatura'].map(temperatura_map)

df_treino['Tempo'] = df_treino['Tempo'].map(tempo_map)

df_teste['Tempo'] = df_teste['Tempo'].map(tempo_map)
```

```python
# Dividir os DataFrames em atributos (X) e rótulos (y)

X_treino = df_treino[['Tempo','Temperatura']]

y = df_treino['JogaFutebol']

X_teste = df_teste[['Tempo','Temperatura']]
```

```python
# Criar e treinar um modelo de árvore de decisão

model = RandomForestClassifier(n_estimators=100)

model.fit(X_treino, y)
```

```python
# Fazer predições no conjunto de teste

df_teste['JogaFutebol'] = model.predict(X_teste)
```

```python
# Mapear as variáveis numéricas de volta em categóricas

tempo_inverso_map = {v: k for k, v in tempo_map.items()}

temperatura_inverso_map = {v: k for k, v in temperatura_map.items()}

df_teste['Tempo'] = df_teste['Tempo'].map(tempo_inverso_map)

df_teste['Temperatura'] = df_teste['Temperatura'].map(temperatura_inverso_map)

print(df_teste)
```
