```python
# imports

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans
```

```python
# Criar dados fictícios

X, _= make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Visualizar os dados

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')

plt.title("Dados Fictícios")

plt.show()
```

```python
# Treinar e aplicar um modelo com base na técnica K-Means

model = KMeans(n_clusters=4)

model.fit(X)

centroids = model.cluster_centers_

labels = model.labels_
```

```python
# Visualizar os clusters e os centróides

plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, linewidths=2, color='red', label='Centroides')

plt.title("Clusters e Centróides Obtidos pelo K-Means")

plt.legend()

plt.show()
```
