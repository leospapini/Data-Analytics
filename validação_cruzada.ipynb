```python
# imports

from sklearn.model_selection import KFold

from sklearn.model_selection import cross_val_score

import pandas as pd

from sklearn.ensemble import RandomForestClassifier
```

```python
# definir conjunto de dados e técnica de aprendizado a ser adotada

df = pd.read_csv('iris.csv')

X=df.drop('Species', axis=1)

y = df['Species']


model = RandomForestClassifier()
```

```python
# validação cruzada

kf = KFold(n_splits=5)

for registros_treino, registros_teste in kf.split(X, y):


    x_treino, x_teste = X.iloc[registros_treino], X.iloc[registros_teste]

    y_treino, y_teste = y.iloc[registros_treino], y.iloc[registros_teste]

# Treinamento

    model.fit(x_treino, y_treino)

# Predição

    y_predito = model.predict(x_teste)

print(x_teste, y_predito)
```
