```python
# imports

import matplotlib.pyplot as plt

import numpy as np

from sklearn.decomposition import PCA
```

```python
# criar conjunto de dados

rng = np.random.RandomState(0)

n_samples = 500

cov = [[3, 3], [3, 4]]

X = rng.multivariate_normal(mean=[0, 0], cov=cov, size=n_samples)
```

```python
# aplicar t√©cnica PCA

pca = PCA(n_components=2).fit(X)
```

```python
# visualizar dados

plt.scatter(X[:, 0], X[:, 1], alpha=0.3, label="Pontos de dados")

for i, (comp, var) in enumerate(zip(pca.components_,

pca.explained_variance_)):
    comp = comp * var
    plt.plot(

        [0, comp[0]],
        [0, comp[1]],
        label=f"Componente {i}",
        linewidth=5,
        color=f"C{i + 2}",

    )

plt.gca().set(

    aspect="equal",
    title="Conjunto de dados bidimensional com componentes principais",
    xlabel="X1",
    ylabel="X2",

)

plt.legend()
plt.show()
```
